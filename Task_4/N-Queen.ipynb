```python
def is_position_safe(placement, row_idx, col_idx, size):
    for prev_row in range(row_idx):
        if placement[prev_row] == col_idx or \
           placement[prev_row] - prev_row == col_idx - row_idx or \
           placement[prev_row] + prev_row == col_idx + row_idx:
            return False
    return True

def find_n_queen_solutions(size, row_idx=0, placement=[], results=[]):
    if row_idx == size:
        results.append(placement[:])
        return
    
    for col_idx in range(size):
        if is_position_safe(placement, row_idx, col_idx, size):
            placement.append(col_idx)
            find_n_queen_solutions(size, row_idx + 1, placement, results)
            placement.pop()
    
    return results

def display_solutions(solutions, size):
    for solution in solutions:
        for row in solution:
            row_representation = ['.'] * size
            row_representation[row] = 'Q'
            print(" ".join(row_representation))
        print("\n")

if __name__ == "__main__":
    try:
        board_size = int(input("Enter the number of queens: "))
        if board_size <= 0:
            raise ValueError("Number of queens must be a positive integer.")
        solutions_found = find_n_queen_solutions(board_size)
        print(f"Total Solutions for {board_size}-Queens: {len(solutions_found)}\n")
        display_solutions(solutions_found, board_size)
    except ValueError as err:
        print(f"Invalid input: {err}")
```
